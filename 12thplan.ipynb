#### I need to map the state names to these given below as our svg map has it that way.

```python
states_str = """kerala
andhra_pradesh
karnataka
maharashtra
gujarat
rajasthan
tamil_nadu
goa
madhya_pradesh
orissa
chhatisgarh
west_bengal
jarkhand
punjab
himachal_pradesh
haryana
uttarakhand
sikkim
uttar_pradesh
bihar
delhi
nagaland
manipur
mizoram
tripura
meghalaya
jammu_and_kashmir
assam
arunachal_pradesh"""
states_arr = states_str.split("\n")
states = []
for state in states_arr:
    temp = str(state)
    temp2 = " ".join(temp.split("_"))
    states.append(temp2)
states.sort()
states
```

```python
close_px = read_csv('datafile.csv', parse_dates=True)
short = []
for i in range(len(close_px.columns)):
    if i%2==1:
        short.append(i)
#print short
temp_df = close_px.ix[:,[0]+short]
temp_df
```

```python
temp_col = temp_df.columns[0]
df1 = temp_df.set_index(temp_col)
df1
```

```python
def matcher(x):
    try:
        return difflib.get_close_matches(x,states)[0]
    except:
        return x.lower()
df1.index = df1.index.map(matcher)
df1
```

```python
filename = "data/factories"
df1.to_csv(filename + ".csv")

s = json.dumps(df1.to_dict())
handle1=open(filename + ".json",'w')
handle1.write(s)
handle1.close();
```
