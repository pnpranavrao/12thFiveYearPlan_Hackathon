```python
states_str = """kerala
andhra_pradesh
karnataka
maharashtra
gujarat
rajasthan
tamil_nadu
goa
madhya_pradesh
orissa
chhatisgarh
west_bengal
jarkhand
punjab
himachal_pradesh
haryana
uttarakhand
sikkim
uttar_pradesh
bihar
delhi
nagaland
manipur
mizoram
tripura
meghalaya
jammu_and_kashmir
assam
arunachal_pradesh"""
states_arr = states_str.split("\n")
states = []
for state in states_arr:
    temp = str(state)
    temp2 = " ".join(temp.split("_"))
    states.append(temp2)
states.sort()
states
```

```python
def matcher(x):
    try:
        return difflib.get_close_matches(x,states)[0]
    except:
        return x.lower()
```

```python
#close_px = read_csv('datafile3.csv', parse_dates=True)
excel_data = pandas.ExcelFile("spout_1112_ds.xls")
sheets = excel_data.sheet_names
close_px = excel_data.parse(sheets[0],parse_dates=True)
#for i in close_px.columns:
#    print str(i)
```

```python
temp_col = close_px.columns[3]
temp_df = close_px.set_index(temp_col)
bool_arr = []
for temp in temp_df.index:
    if "TOTAL" in temp:
        bool_arr.append(True)
    else:
        bool_arr.append(False)
small1 = temp_df[bool_arr]
bool_arr = []
for i in range(len(small1.columns)):
    if i % 2 == 0:
        bool_arr.append(True)
    else:
        bool_arr.append(False)
small2 = small1.ix[:,bool_arr]
small2 = small2.ix[:,2:-2]
small2
```

```python
cols = small2.columns
inds = small2.index
small2 = small2.T
small2
```

```python
def get_states(big_name):
    temp = str(big_name)
    temp = temp.split("-Revised")[0]
    temp = " ".join(temp.split("-"))
    return temp
small2.index = small2.index.map(get_states)
small2.index = small2.index.map(matcher)

def set_column(big_name):
    temp = str(big_name)
    try:
        temp = temp.split("TOTAL -")[1]
    except:
        temp = "Total"
    temp = temp.title()
    return temp
small2.columns = small2.columns.map(set_column)
small2
```

```python
temp_str = "costs"
#Printing to csv
small2.to_csv("data/expenditure/"+temp_str+".csv")
#printing to json
s = json.dumps(small2.to_dict())
handle1=open("data/expenditure/"+temp_str + ".json",'w+')
handle1.write(s)
handle1.close();
```
